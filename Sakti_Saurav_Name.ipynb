```python
# Analyse DTI
dropped_missing['dti_binned'] = pd.cut(dropped_missing.dti, bins=range(0, 30, 1), right=False)
freq_df = dropped_missing.groupby('dti_binned').loan_status.value_counts().unstack()
pct_df = 100*freq_df.divide(freq_df.sum(axis=0), axis=1)
pct_df
# freq_df
pct_df.plot(kind='line', figsize=(15,5))
# Y axis percentage and x axis dti bin
```

```python
# DTI greater than 12 is a alert
```

```python
freq_df = dropped_missing.groupby('addr_state').loan_status.value_counts().unstack()
pct_df = 100*freq_df.divide(freq_df.sum(axis=0), axis=1)
pct_df
# freq_df
pct_df.plot(kind='line', figsize=(20,5))
# Y axis percentage and x axis states
```
